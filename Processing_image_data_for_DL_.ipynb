```python
# important libraries for image processing
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
from PIL import Image
```

```python
# loading an image through matplotlib.image module
img = mpimg.imread('/content/692e6421912e29184674dd58ef9f5e18.jpg')
```

```python
# so our image is rgb image , and basically a 352*563*3 matrix ---> (Height,Width,Channels-(rgb or white))
print('Our image has ', 352*563 , ' no. of pixels ')
```

```python
print(img)
```

```python
# displaying the image from numpy array

img_plot = plt.imshow(img)
plt.show()
```

```python
# Resizing an image using Pillow(PIL) library
img = Image.open('/content/692e6421912e29184674dd58ef9f5e18.jpg')

img_resized = img.resize((350,351))
# (width,height)
```

```python
img_resized.save('resized_cat.jpg')
```

```python
# now diplaying the image from numpy array
img_new = mpimg.imread('/content/resized_cat.jpg')
img_plot = plt.imshow(img_new)
plt.show()
```

```python
# Converting rgb images to greyscale using opencv
import cv2
# imported a computer vision library
img = cv2.imread('/content/692e6421912e29184674dd58ef9f5e18.jpg')
```

```python
type(img)
```

```python
# also a n-dimensional numpy array , also an rgb image
img.shape

# with three channels
```

```python
grayscale_img = cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
```

```python
grayscale_img.shape
```

```python
# that shows its a 1-D array of size (352*563)
```

```python
# cv2.imshow() will display the image in spyder , pycharm etc , but not allowed in Google Colab
from google.colab.patches import cv2_imshow
```

```python
cv2_imshow(grayscale_img)
```

```python
# saving the grascale image //
cv2.imwrite('grayscale_cat.jpg',grayscale_img)
```

```python
# main advantage of grayscaling -> saves the space in big datasets , you can do it when color information is not important
```
